导入相关的库

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from pandas.io.stata import StataReader
stata1 = r"d:/Chrome 下载/计量经济学/salary.dta"
reader1 = StataReader(stata1)
data1=reader1.read()
data1=data1.dropna()
data1
```

进行OLE

```python
y=np.log(data1.salary)
X=sm.add_constant(data1.age)
model= sm.OLS(y,X).fit()
model.summary()
```
